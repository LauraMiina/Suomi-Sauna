# Finnish Sauna Data
Statistics Finland / Tilastokeskus has data about sauna's in Finland and how people are spending time in sauna.
Statistics Finland estimates there to have been over 2.4 million saunas all over Finland at the end of the year 2022. The information about time used in sauna is originally presented as hours.min/day.


```python
import pandas as pd
```

```python
file_path = '/Users/laura.kukkonen/Documents/SuomiSauna/akay_006_201000_20240125-190613.xlsx'
df = pd.read_excel(file_path)
```

```python
df
```

I want to calculate the average time spent in sauna per week.

```python
new_df = pd.DataFrame()
new_df[df.columns[:2]] = df[df.columns[:2]]
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        minutes_per_week_col = column + '_minutes_per_week'
        minutes_per_day_col = column + '_minutes_per_day'
        new_df[minutes_per_week_col] = df[column] * 100 * 7
        new_df[minutes_per_day_col] = df[column] * 100
new_df
```

```python
file_path = "saunadataminutesperweekminutesperday.xlsx"
new_df.to_excel(file_path, index=False)
```

```python
all_rows = new_df[new_df['Gender'] == 'All']

age_columns = all_rows[['Years', '10-14 years_minutes_per_week', '10-14 years_minutes_per_day',
                        '15-24 years_minutes_per_week', '15-24 years_minutes_per_day',
                        '25-44 years_minutes_per_week', '25-44 years_minutes_per_day',
                        '45-64 years_minutes_per_week', '45-64 years_minutes_per_day',
                        '65-_minutes_per_week', '65-_minutes_per_day']]

age_columns
```

```python
all_rows = new_df[new_df['Gender'] == 'All']

age_columns_week = all_rows[['Years', '10-14 years_minutes_per_week',
                        '15-24 years_minutes_per_week',
                        '25-44 years_minutes_per_week',
                        '45-64 years_minutes_per_week',
                        '65-_minutes_per_week']]

age_columns_week
```

```python
age_columns_week = age_columns_week.rename(columns={
    'Years': 'Years',
    '10-14 years_minutes_per_week': '10-14 years',
    '15-24 years_minutes_per_week': '15-24',
    '25-44 years_minutes_per_week': '25-44',
    '45-64 years_minutes_per_week': '45-64',
    '65-_minutes_per_week': '65+'
})

age_columns_week
```

```python
file_path = "saunadataallages.xlsx"
age_columns.to_excel(file_path, index=False)
```

```python
file_path = "saunaagesweek.xlsx"
age_columns_week.to_excel(file_path, index=False)
```

```python
most_popular_place_to_sauna = df2['Where do you sauna?'].value_counts().idxmax()
count_most_common_appearance = df2['Where do you sauna?'].value_counts().max()
print(most_popular_place_to_sauna)
print(count_most_common_appearance)
```
